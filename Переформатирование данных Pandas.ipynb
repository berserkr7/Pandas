# Слияние объектов DF как в базах данных

```python
# Операция слияния или соединения комбинирует наборы данных, соединяя строки по одному или нескольким ключам.
# Эта операция является одной из основных в базах данных. Функция merge в pandas - портал ко всем алгоритмам
# такого рода.
```

```python
df1 = pd.DataFrame({'key' : list('bbacaab'), 'data1' : range(7)})

df2 = pd.DataFrame({'key' : list('abd'), 'data2' : range(3)})
```

```python
df1
```

```python
df2
```

```python
# Это пример соединения МНОГИЕ-К-ОДНОМУ; в объекте df1 есть несколько строк c метками a и b, а в объекте df2
# только одна строка для каждого значения в столбце key. Давай посмотрим, что дает вызов merge для таких объектов:

pd.merge(df1, df2)
```

```python
# Ты заметил, что 'd' пропала? Потому что ее не было 
```

```python
# Обрати внимание, что я не указал, по какому столбцу производить соединения. В таком случае merge использует
# в качестве ключей столбцы с одинаковыми именами. Однако, все же рекомендуется указывать столбцы явно:

pd.merge(df1, df2, on='key')
```

```python
# Если имена столбцов в объектах различаются, то можно задать их порознь:

df3 = pd.DataFrame({'lkey' : list('bbacaab'), 'data1' : range(7)})

df4 = pd.DataFrame({'rkey' : list('abd'), 'data2' : range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
# Вероятно, вы уже обратили внимание, что значения 'c' и 'd' и ассоцированные с ними данные отсутствуют в 
# результатирующем объекта. По умолчанию функция merge производит внутренне соединение ('inner'); в результатирующий
# объект попадают только ключи, присутствующие в обоих объектах-аргументах. Альтернативы соединения это
# 'left', 'right', 'outer'. В случае внешнего соединения 'outer' берется объдинение ключей, то есть получается все
# тоже самое, что если использовать левое и правое объединение ключей одновременно

pd.merge(df1, df2, how='outer') # то есть способ объединения коючей задаем через параметр how!!!!!
```

```python
# Рассмотрим различные типв соединения, задаваемые аргументов how:
#
# 'inner' - брать только комбинации ключей, встречающихся в обеих таблицах
# 'left' - брать все ключи, встречающиеся в левой таблице
# 'right' - брать все ключи, встречающиеся в правой таблице
# 'outer' - брать все комбинации ключей
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
# Теперь мы переходим к слиянию типа МНОГИЕ-КО-МНОГИМ, поведение которого корректно определено, хотя неочевидно на первый взгляд
```

```python
df1 = pd.DataFrame({'key' : list('bbacab'), 'data1' : range(6)})

df2 = pd.DataFrame({'key' : list('ababd'), 'data2' : range(5)})
```

```python
df1
```

```python
df2
```

```python
# Ну что, давай попробуем объединить)))
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
# Соединение МНОГИЕ-КО-МНОГИМ порождает декартово произведение строк. Поскольку в левом объекте было три 
# строки с ключом 'b', а в правом две, то в результатирующем объекте таких строк получилось шесть! Метод соединения
# оказывает влияние только на множество различных ключей в результате:

pd.merge(df1, df2, how='inner')
```

```python
# Для соединения по нескольким ключам следует передать список имен столбцов:

left = pd.DataFrame({'key1' : ['foo', 'foo', 'bar'], 'key2' : ['one', 'two', 'one'], 'lval':[1,2,3]})

right = pd.DataFrame({'key1' : ['foo', 'foo', 'bar', 'bar'], 'key2' : ['one', 'one', 'one', 'two'], 'rval':[4,5,6,7]})
```

```python
# Последний момент, касающийся операции слияния, - обработка одинаковых имен столбцов, у функции merge имеется
# параметр suffixes, позволяющий задать строки, которые должны дописываться в конец одинаковых имен в левом и правом
# объектах DataFrame:
```

```python
left
```

# Соединение по индексу

```python
# Иногда ключ или ключи соединения находятся в индексе объекта. То есть жо этого выше я рассмотрел, когда ключ находится
# просто в столбце. В таком случае можно задать параметр left_index=True или right_index=True (или то и другое), чтобы
# указать, что в качестве ключа соединения следует использовать индекс:

left1 = pd.DataFrame({'key' : list('abaabc'), 'value' : range(6)})

right1 = pd.DataFrame({'group_val' : [3.5,7]}, index=list('ab'))
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
# По умолчанию соединение производится по пересекающимся ключам, но можно вместо пересечения выполнить объединение
# указав внешнее соединенение:

pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
# В примере выше на другой оси нет перекрытия, и она, как видим, является отсортированным объединением (внешним соединением)
# индексов. Но можно образовать и пересечение индексов, если передать параметр join='inner'


s4 = pd.concat([s1, s3])
s4
```

```python
# Это и называется ДЛИННЫМ форматом для нескольких временных рядов или других наблюдений с двумя и более
# ключами (в данном случае ключами являются дата и показатель item). Каждая строка таблицы соответствует одному 
# наблюдению.
#
# Так данные часто хранятся в реляционных БД типа MySQL.
#
# Чтобы в таблице задать новый столбец с другими значениями индексов, используйте в Pandas метод pivot. 
# Для изменения формы указываются аргументы: index (индекс), columns (столбцы), values (значения). 
# Вам может понадобится pivot тогда, когда нужно зафиксировать новый индекс с новым столбцом.

pivoted = ldata.pivot(index='date', columns='item', values='value') # то есть значения item будут новыми столбцами
pivoted
```

```python
# Отметим, что метод pivot это не более чем сокращенный способ создания иерархического индекса
# с помощью  set_index и последующего вызова unstack():

unstacked = ldata.set_index(['date', 'item']).unstack('item')
unstacked
```

# Поворот из широкого в длинный формат

```python
# Обратной к pivot операцией является pd.melt. Вместо того, чтобы преобразовывать один столбец в несколько
# в новом объекте DF, она объединяет неколько столбцов в один, порождая DataFrame длинне входного.



df = pd.DataFrame({'key' : ['foo', 'bar', 'baz'], 'A' : [1, 2, 3], 'B' : [4, 5, 6], 'C' : [7, 8, 9]})
df
```

```python
# Столбец key может быть индикатором группы, а остальные столбцы - значениями данных. При использовании функции
# pd.melt необходимо указать какие столбцы являются индикаторами группы (если таковые имеются). Будем считать,
# что в даннм случае 'key' - единственный индикатор группы:

melted = pd.melt(df, ['key'])
melted

# melt переводится как "расплавленный", что логично, таблица становится больше вниз будто таит
# pivot переводится как "вращаться"
```

```python
# Применив pivot можем вернуться к исходной форме, то есть melt и pivot антагонисты друг друга

reshe = melted.pivot(index='key', columns='variable', values='value')
```

```python
# Поскольку в результате работы pivot из столбца создается индекс, используемый как метки строк, нам понадобится
# использовать reset_index, чтобы восстановить из индекса столбец:


reshe.reset_index()
```

```python
# Функция melt или метод DataFrame.melt предназначены 
# для преобразования одного или более столбцов в строки с измеряемыми значениями. 
# По умолчанию новый столбец с полученными строками имеет название variable, 
# а столбец с измеряемыми значениями — value. Эти названия могут быть изменены параметрами var_name и value_name.
```
