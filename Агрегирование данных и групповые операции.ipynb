# Метод apply: часть общего принципа разделение-применение-объединение

```python
# Предположим, что в чаевых надо выбрать первые 5 значений tip_pct в каждой группе:
tips
```

```python
def top(df, n=5, column='tip_pct'): # сортирует столбец
    return df.sort_values(column)[-n:]

top(tips, n=6)
```

```python
# Попробуем теперь сгруппировать по столбцу smoker и вызывать метод apply, передав ему эту функцию:

tips.groupby('smoker').apply(top)
```

```python
# Если передать методу apply функцию, которая применяет позиционные аргументы или именованные, то их можно
# передать вслед за самой функций:

tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

# Квантильный и интервальный анализы

```python
df = pd.DataFrame({'data1' : np.random.randn(1000), 'data2' : np.random.randn(1000)})
```

```python
df
```

```python
df.groupby('data1').agg('min', 'max')
```

```python
factor = pd.cut(df.data1, 5)
factor
```

```python
# Объект Categorial, возвращаемый функцией cut, можно передать непосрественно groupby. Следовательно, 
# набор статистик для столбца data2 можно вычислить следующим образом:

def get_stats(group):
    return {'min' : group.min(), 'max' : group.max(), 'count' : group.count()}
```

```python
grouped = df.data2.groupby(factor)
grouped
```

```python
grouped.apply(get_stats).unstack()
```

```python
# Это были интервалы одинаковой длины, а чтобы вычслить интервалы равного размера на основе выборочных квантилей,
# нужно использовать функцию qcut. Параметр labels отвечает за получение номеров квантилей:
```

```python
gr = pd.qcut(df['data1'], 10)
gr
```

```python
gr = pd.qcut(df['data1'], 10, labels=False)
gr
```

```python
df.data2.groupby(gr).apply(get_stats).unstack()
```

# Сводные таблицы и перекрестное табулирование

Сводная таблица - это средство обощения данных, оно агрегирует таблицу по одному или нескольим ключам и строит
другую таблицу, в которой одни групповые ключи расположены в сроках, а другие в столбцах. Бибилиотека pandas
позволяет строить сводные таблицы с помощью механизма groupby в сочетании с параметрами изменения формы. В классе
DF имеется метод pivot_table, которая помимо groupby может еще добавлять частичные итоги, которые называются
маргиналами.

```python
tips
```

```python
tips.groupby(['day', 'smoker']).mean()
```

```python
tips.pivot_table(index=['day', 'smoker'])
```

```python
tips
```

```python
?pd.pivot_table
```

```python
# Пусть теперь требуется агрегировать только столбцы tip_pct и size, добавив еще группировку по time.
# Поместим средние по smoker в столбцы таблицы, а по day в строки:

tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'], columns='smoker')

```

```python
tips.pivot_table(index=['time', 'day'], columns=['smoker'])
```

```python
# Эту таблицу можно было бы дополнить, включив частичные итоги с помощью параметра margins=True. Тогда будут добавлены
# строка и столбце с меткой All, значениями в которых будут групповые статистики по всем данным на одном уровне:
```

```python
tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'], columns='smoker', margins=True)

# Здесь столбцы All содержат СРЕДНИЕ значения без учета являются или не являются курящими. А строка с All 
# считает средние для курящих и некурящих
```

```python
# Для применения другой функции агрегирования нужно передать параметр aggfunc момимо параметр margins(он лишь добаляет вывод)
```

```python
tips.pivot_table(['tip_pct'], index=['time', 'smoker'], columns='day', margins=True, aggfunc=len)
```

```python
# Для восполнения пустых значений можно воспользоваться параметром fill_value
```

```python
tips.pivot_table(['tip_pct'], index=['time', 'smoker'], columns='day', margins=True, aggfunc=len, fill_value=0)
```

```python
# Сводка аргументов метода pivot_table:
    
# values - имя одного или нескольких столбцов, по которым производится агрегирование. По умолчанию по всем столбцам

# index - имена столбцов или другие групповые ключи для группировки по строкам результатирующей сводной таблицы

# columns - имена столбцов или другие групповые ключи для группировки по столбцам результатирующей сводной таблицы

# aggfunc - функция агрегирования или список таких функций, по умолчанию mean. Можно задать произвольную функцию

# fill_value - чем заменить отсутствующие значения в таблице

# dropna - если True, то не включать столбец в котором все значения отсутствуют

# margins - добавлять некоторую сводку итоги
```

# Таблицы сопряженности

Таблица сопряженности, или перекрестная таблица (кросс таблица) - частный случай сводной таблицы, в которой представлены групповые частоты.

```python
data = pd.DataFrame({'sample' : list(range(1,11)), 'Национальность' : ['USA', 'Japan', 'USA', 'Japan', 'Japan', 'Japan', 'USA', 'USA', 'Japan','USA'], 'Рука' : list('ПРППРППРПП')})
```

```python
data
```

```python
# К примеру, мы хотим обобщить эти данные по национальности и рукам. Для этих целей можно использотвать pivot_table
```

```python
data.pivot_table( index=[ 'Национальность'], columns='Рука',  aggfunc='count')
```

```python
# Но удобнее это сделать с помощью кросс-таблицы: pandas.crosstab
```
